```python
import pandas as pd
```

```python
bear_df1 = pd.read_csv('bear_attacks(kaggle).csv')
```

```python
bear_df1.head()
```

```python
bear_df1_dropped_coloumns = bear_df1.drop(['Details', 'Name', 'Age', 'Gender'], axis =1 )
```

```python
black_brown_bears_df1 = bear_df1_dropped_coloumns[bear_df1_dropped_coloumns['Bear'] != 'Polar']
```

```python
black_brown_bears_df1.head()
```

```python
canadian_regions = ['Alberta', 'British Columbia', 'Manitoba', 'New Brunswick', 'Newfoundland and Labrador', 'Northwest Territories', 'Nova Scotia', 'Nunavut', 'Ontario', 'Prince Edward Island', 'Quebec', 'Saskatchewan', 'Yukon']
df_bears1_canada = black_brown_bears_df1[black_brown_bears_df1['Location'].str.contains('|'.join(canadian_regions), case=False, na=False)]
```

```python
df_bears1_canada.head()
```

```python
df_bears1_canada.to_excel('bear_attacks_canada1.xlsx', index=False)
```

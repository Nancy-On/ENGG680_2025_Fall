## Part 6: Linear LSE
### (35 marks)

### Curve Fitting

curvefitting.png

### Observations: values of a function f(a) in range of - to , contaminated with random noise

### Unknown States: coefficients of polynomial functions for curve fitting

```python
# step 1
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# step 2
# Load data.xlsx file, which contains the observations (with noise), variance of observations and true observations
# hint: use pandas library to parse excel file and use numpy library to convert it to array
# (1 mark)
df_data = pd.read_excel('data.xlsx')
arr_data = df_data.to_numpy()

#store the noisy observations, variance of observations and true observations into numpy arrays seperately
# (1 mark)
l = arr_data[:,0]
l_var = arr_data[:,1]
l_true = arr_data[:,2]
```

```python
# step 3
# define the number of observation n
# (1 mark)
n = len(l)


# define the observation weight matrix P for LSE as identity matrix
# identity weight matrix simply means the observations will be unweighted in the following LSE
# (1 mark)
P = np.identity(n)

# define the range of function input(i.e., constants a), from start to end; and the numerical array of a
start = -1* np.pi
end = np.pi
a = np.linspace(start, end, n)
```

```python
# step 4
# use the matplotlib to plot the true and noisy observations together in one single figure
# use scatter for noisy observation plotting
# for the true observation, label it as "f(a)"" and set color as 'k'
# for the noisy observation scatter plot, label it as "f(a) + e", set color as 'tab:blue', set marker as 'o' and set alpha as 0.3
# include proper axis labels, legend and title
# scale the range of x-axis to be -1 to 1
# keep this format for the rest of the plots in this lab assignment
# (2 marks)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))

#creating plots for true and noisy observations
ax.plot (a, l_true, label='f(a)', color= 'k')
ax.scatter(a, l, label='f(a) + e', color='tab:blue', marker='o',alpha =0.3)

#Adding labels and adjusting axes
ax.set_xlabel ("a")
ax.set_ylabel ("Observations")
ax.set_title ("Plot of True and Noisy Observations")
ax.legend (loc='best')
ax.set_xlim (-1,1)


ax.grid(True, which='both',ls='-',color='0.65')
fig.tight_layout()
```

```python
# step 5
# define a curve fitting function named fit_polynomial_curve
# the function should implement the linear least square estimation
# the function should include all important intermediate steps in LSE
# the input and output of the funciton are described in the comment below in red.
# please keep code clean and well-commented
# (7 marks)

def fit_polynomial_curve(a: np.ndarray, l: np.ndarray, P: np.ndarray, deg: int) -> tuple:
    '''
    Args:
        a:              function input (constants)
        l:              noisy observations
        P:              observation weight matrix
        deg:            degree of polynomial to fit to data
    Returns:
        x:              estimated polynomial coefficients
        cov_x:          uncertainty in estimated coefficients
        apv:            a posterier variance factor (goodness of fit)
        v:              observation residuals
    '''

    #defining general matrix A for polynomial
    A = np.vander(a, N= deg + 1, increasing = True)

    #computing normal matrix A^TPA and A^TPl
    ATPA = A.T @ P @ A
    ATPl = A.T @ P @ l

    #solving system of normal equations
    x= np.linalg.solve(ATPA, ATPl)

    #Finding the equation for the line of best-fit
    y=np.poly1d(x[::-1])

    #Computing vector of residuals
    v = l - (A @ x)

    #Computing posterior variance factor
    degrees_of_freedom = n-(deg+1)
    apv = (v.T @ P @ v)/ degrees_of_freedom

    #Computing the covariance of estimated coefficients
    cov_x = apv * np.linalg.inv(ATPA)

    return (x, cov_x, apv, v)
```

```python
# step 6
# call the fit_polynomial_curve function and fetch the results
# use polynomial degrees of: 2, 4, 6
# (1 marks)

degree = [2,4,6]
results = {}

for deg in degree:
    x, cov_x, apv, v = fit_polynomial_curve(a, l, P, deg)
    results[deg]= {'x':x,
                   'cov_x' : cov_x,
                   'apv': apv,
                   'v':v }

    print(f"Results for polynomial with degree {deg}:")
    print(f"Estimated polynomial coefficients: {results[deg]['x']}")
    print(f"Uncertainty in estimated coefficients: {results[deg]['cov_x']}")
    print(f"Posterior variance factor: {results[deg]['apv']}")
    print(f"Observation residuals: {results[deg]['v']}\n")
```

```python
# step 7
# use the estimated coefficients x and constants array a, construct polynomial functions for each degree
# hint: use polyval function from numpy library (check input argument carefully)
# (1 marks)

quadratic_uw = np.polyval(results[2]['x'][::-1], a)
quartic_uw = np.polyval(results[4]['x'][::-1], a)
hexic_uw = np.polyval(results[6]['x'][::-1], a)

```

```python
# step 8
# use the matplotlib to plot each polynomial function, noisy observations l, true observations l_true, together in one single figure
# for the polynomials of line, cubic and quintic, use color of orange, green and red respectively.
# (2 marks)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))
ax.plot (a, l_true, label='f(a)', color= 'k')
ax.scatter(a, l, label='f(a) + e', color='tab:blue', marker='o',alpha =0.3)
ax.plot (a, quadratic_uw, label='Quadratic function', color='orange')
ax.plot (a, quartic_uw, label='Quartic function', color='green' )
ax.plot (a, hexic_uw, label ='Hexic function', color='red')
ax.set_xlabel ("a")
ax.set_ylabel ("Observations")
ax.set_title ("Plot of True and Noisy Observations with Quadratic, Quartic and Hexic Functions")
ax.legend (loc='best')
ax.set_xlim (-1,1)


ax.grid(True, which='both',ls='-',color='0.65')

fig.tight_layout()
```

```python
# step 9
# print the a posterier variance factor for each polynomial function
# (3 marks)
print("Posterior Variance Factor for Each Polynomial:")
for deg in degree:
    print(f"Posterior variance factor for degree {deg} = {results[deg]['apv']}")
```

```python
# step 11
# comment on the results of a posterier variance factor and norm, what pattern do you observe
# (2 marks)

# From the calculated posterior variance factor and norm value, it can be observed that as the degree of the polynomial increases, the norm and posterior variance factor decreaase.
# The norm represents the residuals in curve fitting. A smaller norm means that residuals are smallers, indicating that the true oberservations are close to the predicted values.
# The quadratic function has the largest norm and while the hexic function has the smallest norm.
# The posterior variance factor is a measure of the goodness of fit in curve fittting. The smaller the values, the smaller the variance of the measurements.
# The quadratic function has the largest posterior variance factor while the hexic function has the smallest.

# When both norm and posterior variance factor are taken into account, we can determine which model best fits our data.
# In this case, as the degree of the polynomial increases, the fit of the curve improves, meaning that the model is able to accurately describe the pattern in our dataset, residuals are small and estimated coefficients are reliable.
# Continuing to increase the degree will eventually lead to a point where improvements are small. When this occurs, there is a risk of overfitting, which is when the model becomes too complex for the dataset, leading to high variance and poor generalization
```
